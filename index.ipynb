# 1. Introduction
## Business Problem Understanding

**Context** 

Olist adalah marketplace yang menawarkan solusi SaaS kepada berbagai penjual. Beberapa penjual menunjukkan performa penjualan yang rendah, terutama di kategori produk tertentu. Manajemen ingin membantu penjual ini meningkatkan penjualan mereka dengan memberikan insight berbasis data dan rekomendasi strategi bisnis yang lebih baik. Target akhirnya adalah meningkatkan penjualan penjual tersebut untuk meningkatkan revenue Olist secara keseluruhan.

Selain meningkatkan penjual yang menunjukkan performa penjualan yang rendah, Manajemen juga ingin membantu dalam peningkatan penjualan untuk semua seller, dengan memberikan rekomendasi terkait barang yang umumnnya dijual didaerah/kota seller berada beserta rata-rata harga per barangnya. 

**Problem Statement**

Beberapa penjual di marketplace Olist memiliki tingkat penjualan yang rendah dan kontribusi revenue yang tidak optimal. Ada berbagai faktor yang mungkin memengaruhi kinerja mereka, seperti harga produk, ulasan pelanggan, metode pembayaran yang ditawarkan, dan strategi pemasaran yang tidak efektif. Olist ingin memahami faktor-faktor ini dan menerapkan strategi berbasis data untuk membantu penjual meningkatkan performa penjualan mereka.

    - Bagaimana perkembangan pertumbuhan penjualan Olist selama periode waktu tertentu?
    - Kategori produk atau penjual mana yang mendorong volume penjualan tertinggi?
    - Apakah terdapat perbedaan signifikan dalam kinerja penjualan di berbagai wilayah di Brasil?
    - Bagaimana Olist dapat menggunakan wawasan ini untuk meningkatkan strategi bisnis di masa mendatang?

**Goals**

Tujuan dari analisis ini adalah:

1. Jangka Pendek: Mengidentifikasi faktor-faktor yang menyebabkan rendahnya performa penjual dan memberikan rekomendasi yang terukur untuk meningkatkan penjualan mereka.

2. Jangka Panjang: Membantu penjual dengan performa rendah & penjual baru meningkatkan sales growth rate sebesar 20% dalam 3 bulan berikutnya dan meningkatkan sales volume berdasarkan kategori produk, penjual, dan wilayah pelanggan.

*KPI:*

- Peningkatan Sales Growth Rate untuk penjual yang berperforma rendah (khususnya) dan penjual secara keseluruhan (umumnya)
- Peningkatan Sales Volume berdasarkan kategori produk, penjual, dan wilayah pelanggan.

**Evaluation Metrics**

1. Identifikasi Penjual Berperforma Rendah 

    Dari data penjualan Olist, identifikasi penjual dengan Sales Growth Rate negatif dalam 3 bulan terakhir. Misalnya, penjual yang mengalami penurunan penjualan sebesar -15% 

2. Analisis Produk

    Analisis produk yang dijual oleh penjual berperforma rendah. Identifikasi produk dengan jumlah ulasan negatif yang signifikan, harga produk yang terlalu tinggi dibandingkan dengan produk serupa, dan kategori produk yang kurang diminati pelanggan.

3. Rencana Intervensi

Setelah menganalisis faktor-faktor tersebut, Olist memberikan rekomendasi:
    - Harga: Menurunkan harga produk yang terlalu tinggi untuk meningkatkan daya saing dan memanfaatkan strategi diskon 
             pada produk tertentu.

    - Deskripsi Produk: Memperbarui deskripsi produk untuk mencerminkan manfaat dan fitur produk dengan lebih baik.

    - Ulasan: Mengikuti program peningkatan kualitas dengan memberikan saran kepada penjual untuk meningkatkan 
              produk berdasarkan feedback pelanggan.

    - Promosi: Menggunakan kampanye pemasaran yang menargetkan segmen pelanggan tertentu dengan promosi diskon.

4. Pelaksanaan Rekomendasi

    Olist menerapkan strategi yang telah direkomendasikan kepada penjual. Harga beberapa produk diturunkan, dan kampanye diskon selama 2 minggu dilakukan untuk mendorong pembelian.

5. Evaluation Metrics (Measurable)
> Sales Growth Rate: Setelah penerapan strategi baru, hitung Sales Growth Rate selama 3 bulan:
    
$$ 
\text{Sales Growth Rate} = \frac{\text{Penjualan Saat Ini} - \text{Penjualan Sebelumnya}}{\text{Penjualan Sebelumnya}} \times 100\% 
$$

Jika sebelumnya penjualan bulanan adalah $10.000 dan setelah intervensi menjadi $12.500, maka:

$$ 
\text{Sales Growth Rate} = \frac{\text{12.500} - \text{10.000}}{\text{10.000}} \times 100\% = 25\%
$$

Dengan hasil ini, penjual telah mencapai target peningkatan 20% dalam 3 bulan.

> Sales Volume : Menghitung Sales Volume berdasarkan kategori penjual/seller.
   
    - Sales Volume per Penjual:

$$
\text{Sales Volume per Penjual} = \sum (\text{Jumlah Unit Terjual per Penjual})
$$


Metrik ini akan menangkap total jumlah unit yang terjual selama periode tertentu. Dengan memecah volume penjualan berdasarkan kategori penjual, kita dapat mengidentifikasi area yang berkinerja baik dan yang memerlukan peningkatan.

## **2. Geolocation Dataset Cleaning**

Tables used:
1. Geolocation

Output DataFrame Variable : `geolocation`

```python
# Normalizing Cities Name Formats
df_geo.geolocation_city = df_geo.geolocation_city.apply(lambda x: unidecode.unidecode(x))
```

```python
df_geo
```

```python
df_geo_zip_codes = df_geo['geolocation_zip_code_prefix'].unique()

df_zip_code_desc = pd.DataFrame(columns = ['Zip Code', 'Min Lat', 'Max Lat', 'Lat Diff', 'Median Lat',
                                           'Min Lng', 'Max Lng', 'Lng Diff', 'Median Lng',
                                           'Count', 'State', 'City'])

geo_list = []

def compute_geo_stats(df):
    min_lat = df['geolocation_lat'].min()
    max_lat = df['geolocation_lat'].max()
    median_lat = df['geolocation_lat'].median()
    min_lng = df['geolocation_lng'].min()
    max_lng = df['geolocation_lng'].max()
    median_lng = df['geolocation_lng'].median()
    
    return min_lat, max_lat, median_lat, min_lng, max_lng, median_lng

for zip_code in df_geo_zip_codes:
    df_temp = df_geo[df_geo['geolocation_zip_code_prefix'] == zip_code]
    states = df_temp.geolocation_state.unique()
    
    for state in states:
        df_temp_states = df_temp[df_temp['geolocation_state'] == state]
        cities = df_temp_states.geolocation_city.unique()
        
        for city in cities:
            df_temp_states_city = df_temp_states[df_temp_states['geolocation_city'] == city]
            min_lat, max_lat, median_lat, min_lng, max_lng, median_lng = compute_geo_stats(df_temp_states_city)
            geo_list.append({
                'Zip Code': zip_code,
                'Min Lat': min_lat,
                'Max Lat': max_lat,
                'Lat Diff': max_lat - min_lat,
                'Median Lat': median_lat,
                'Min Lng': min_lng,
                'Max Lng': max_lng,
                'Lng Diff': max_lng - min_lng,
                'Median Lng': median_lng,
                'Count': len(df_temp_states_city),
                'State': df_temp_states_city.geolocation_state.unique()[0],
                'City': df_temp_states_city.geolocation_city.unique()[0]
            })
    
df_zip_code_desc = pd.concat([df_zip_code_desc, pd.DataFrame(geo_list)], 
                             ignore_index = True)

display(df_zip_code_desc)
```
